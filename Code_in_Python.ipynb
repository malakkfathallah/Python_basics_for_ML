# Gaussian distribution

This example generates a plot of an idealized Gaussian distribution. The x-axis are
 the observations and the y-axis is the likelihood of each observation. In this case, observations
 around 0.0 are the most common and observations around-3.0 and 3.0 are rare or unlikely.

```python
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm

x_axis = arange(-3 , 3, 0.001)

y_axis = norm.pdf(x_axis , 0 , 1)

pyplot.plot(x_axis , y_axis)
pyplot.show()
```

This code generates a specified number of random numbers (e.g. 10,000) drawn from a Gaussian distribution with a mean of zero and a standard deviation of 1.
We can then scale these numbers to a Gaussian of our choosing by rescaling the numbers. This can be made consistent by adding the desired mean (e.g. 50) and multiplying the value by the standard deviation (e.g. 5).
 In the case of this dataset, we do have enough data and the plot is blocky because the
 plotting function chooses an arbitrary sized bucket for splitting up the data.

```python
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot
seed(1)
data = 5 * randn(10000) + 50
pyplot.hist(data)
pyplot.show()
```

**When to Seed the Random Number Generator?** <br>  There are times during a predictive modeling project when you should consider seeding the
 random number generator. Lets look at two cases: <br>
 1. **Data Preparation.** Data preparation may use randomness, such as a shu e of the data
 or selection of values. Data preparation must be consistent so that the data is always
 prepared in the same way during tting, evaluation, and when making predictions with
 the nal model.
 2.  **Data Splits.** The splits of the data such as for a train/test split or k-fold cross-validation
 must be made consistently. This is to ensure that each algorithm is trained and evaluated
 in the same way on the same subsamples of data

**Why Controlling for Randomness Matters?** <br>
Lets say you have a stochastic algorithm  which means it uses randomness internally, like shuffling data, random weight initialization, or sampling data points.<br>

Even if you train it on the exact same data, you may get slightly different accuracy each time you run it. <br>

Thats normal! But now the question becomes:<br>

How can I fairly report how well my model performs? <br>

If you only test it once, your result might be lucky or unlucky (high or low). Thats not reliable.


**Solution** : Evaluate the model multiple times with different randomness to understand how stable it really is. <br>
Types of uncertainty:
1. Data uncertainty: We repeat the training on different train/test splits (e.g., via K-Fold Cross Validation) -> Variations in performance due to different train/test splits
to measure how the performance changes due to data split differences. <br>
2. Algorithm uncertainty: We run the model multiple times on the same data split, to see how its own randomness affects results. ->	Variations due to randomness in the model itself

**Best Practice:** Combine Both Types of Uncertainty


* Change the train/test splits each time (to account for data uncertainty)

* Allow randomness in the model each time (to account for algorithm uncertainty)

* Do this process 30 or more times

* Then calculate the mean, standard deviation, or even confidence intervals of the accuracy (or other metric)

# Confidence Interval:

A confidence interval is a bounds on the estimate of a population variable. It is an interval
 statistic used to quantify the uncertainty on an estimate

EX.  Consider a model with an error of 20%, or 0.2 (error = 0.2), on a validation dataset with 50
 examples (n = 50). We can calculate the 95% condence interval (z = 1.96) as follows

```python
# binomial confidence interval
from math import sqrt
# calculate the interval
interval = 1.96 * sqrt( (0.2 * (1- 0.2)) / 50)
print('%.3f' % interval)
```

We can then make claims such as:<br>
* The classification error of the model is 20% +/- 11% <br>

* The true classification error of the model is likely between 9% and 31%.




 The **proportion.confint()** Statsmodels function an implementation of the binomial pro
portion con dence interval. By default, it makes the Gaussian assumption for the Binomial
 distribution, although other more sophisticated variations on the calculation are supported.

 **Function Arguments:**
  * the count of successes (or failures)
  * the total number of trials
  * the significance level
<br>**Function Returns:**
  * returns the lower and upper bound of the confidence interval.<br>The
 example below demonstrates this function in a hypothetical case where a model made 88 correct
 predictions out of a dataset with 100 instances and we are interested in the 95% condence
 interval (provided to the function as a signi cance of 0.05)

```python
# calculate the confidence interval
from statsmodels.stats.proportion import proportion_confint
# calculate the interval
lower, upper = proportion_confint(88, 100, 0.05)
print('lower=%.3f, upper=%.3f' % (lower, upper))
```

**Nonparametric Confidence Interval:**
<br>  Often we do not know the distribution for a chosen performance measure. Alternately, we may
 not know the analytical way to calculate a confidence interval for a skill score.
 <br> In these cases, the bootstrap resampling method can be used as a nonparametric method for
 calculating confidence intervals, nominally called bootstrap confidence intervals.

```python
from numpy.random import rand ,randint
from numpy import mean,median
#generate dataset
dataset = 0.5 + rand(1000) * 0.5
#bootstrap
scores = list()
for _ in range(100):
  #bootstrap sample
  i = randint(0,1000,1000)
  sample = dataset[i]
  #calculate and store statistic
  statistic = mean(sample)
  scores.append(statistic)
# calculate the median
print('median=%.3f' % median(scores))
```

```python
from numpy import percentile
# calculate 95% confidence intervals (100- alpha)
alpha = 5.0
# calculate lower percentile (e.g. 2.5)
lower_p = alpha / 2.0
# retrieve observation at lower percentile
lower = max(0.0, percentile(scores, lower_p))
# calculate upper percentile (e.g. 97.5)
upper_p = (100- alpha) + (alpha / 2.0)
# retrieve observation at upper percentile
upper = min(1.0, percentile(scores, upper_p))
print('%.1f confidence interval %.1f and %.1f' % (alpha, lower*100, upper*100))
```

# Prediction Interval:

 A prediction interval is a quanti cation of the uncertainty on a prediction. <br>It provides a
 probabilistic upper and lower bounds on the estimate of an outcome variable.

Ex. Prediction Interval Calculation for Linear Regression Model

```python
# generate related variables
from numpy import mean
from numpy import std
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot
# seed random number generator
seed(1)
# prepare data
x = 20 * randn(1000) + 100
y = x + (10 * randn(1000) + 50)
# summarize
print('x: mean=%.3f stdv=%.3f' % (mean(x), std(x)))
print('y: mean=%.3f stdv=%.3f' % (mean(y), std(y)))
# plot
pyplot.scatter(x, y)
pyplot.show()
```

```python
from scipy.stats import linregress
# fit linear regression model
b1, b0, r_value, p_value, std_err = linregress(x, y)
print('b0=%.3f, b1=%.3f' % (b1, b0))
# make prediction
yhat = b0 + b1 * x
# plot data and predictions
pyplot.scatter(x, y)
pyplot.plot(x, yhat, color='r')
pyplot.show()
```

 We are now ready to make a prediction with our simple linear regression model and add a
 prediction interval. We will t the model as before. This time we will take one sample from
 the dataset to demonstrate the prediction interval. We will use the input to make a prediction,
 calculate the prediction interval for the prediction, and compare the prediction and interval to
 the known expected value.

```python
# define the prediction
x_in = x[0]
y_out = y[0]
yhat_out = yhat[0]
#estimate stdev pf yhat
sum_errs = sum((y - yhat)**2)
stdev = sqrt(1/(len(y)-2) * sum_errs)
```

We will use the significance level of 95%, which corresponds to 1.69 standard deviations
 from the Gaussian distribution.

```python
# calculate prediction interval
interval = 1.96 * stdev
lower, upper = yhat_out- interval, yhat_out + interval
```

```python
print('Prediction Interval: %.3f' % interval)
print('95%% likelihood that the true value is between %.3f and %.3f' % (lower, upper))
print('True value: %.3f' % yhat_out)
# plot dataset and prediction with interval
pyplot.scatter(x, y)
pyplot.plot(x, yhat, color='red')
pyplot.errorbar(x_in, yhat_out, yerr=interval, color='black', fmt='o')
pyplot.show()
```

B. Statistical Tests:
<br> Before you can apply the statistical tests, you must know how to interpret the results.
<br>Each
 test will return at least two things:
 1. **Statistic**: A quantity calculated by the test that can be interpreted in the context of the
 test via comparing it to critical values from the distribution of the test statistic.
 2. **p-value**: Used to interpret the test, in this case whether the sample was drawn from a
 Gaussian distribution.

 Each test calculates a test-specific statistic.<br>
  This statistic can aid in the interpretation of the
 result, although it may require a deeper pro ciency with statistics and a deeper knowledge of  the specific statistical test.<br>
  Instead, the p-value can be used to quickly and accurately interpret
 the statistic in practical applications.<br>
  The tests assume that the sample was drawn from a
 Gaussian distribution. <br>Technically this is called the null hypothesis, or H0. A threshold level is
 chosen called alpha, typically 5% (or 0.05), that is used to interpret the p-value. <br>In the SciPy
 implementation of these tests, you can interpret the p value as follows.

 * **p-value  <= alpha**: significant result, reject null hypothesis, not Gaussian (H1).
 *  **p-value > alpha**: not significant result, fail to reject null hypothesis, Gaussian (H0).


Types of statistical tests:
1. Shapiro-Wilk Test
2. D'Agostino's K^2 Test
3. Anderson-Darling Test

 The **Shapiro-Wilk test** evaluates a data sample and quanti es how likely it is that the data
 was drawn from a Gaussian distribution.<br>
 It is believed to be a reliable test of normality, although there is
 some suggestion that the test may be suitable for smaller samples of data.<br>
  The function returns both the W-statistic calculated by the test and the p-value.

```python
#example of the shapiro-wilk test
from numpy.random import seed , randn
from scipy.stats import shapiro
#seed the random number generator
seed(1)
#generate univariate observations
data = 5 * randn(100) + 50
#normality test
stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
#interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

The **D'Agostino's K2 test** calculates summary statistics from the data, namely **kurtosis** and
 **skewness**, to determine if the data distribution departs from the normal distribution.<br>
 1. Skew -> is a quantification of how much a distribution is pushed left or right, a measure of asymmetry in the distribution.
 2. Kurtosis -> is a quantification of how much of the distribution is
 in the tail.

```python
#example of the d'agostino and pearson's test
from numpy.random import seed, randn
from scipy.stats import normaltest
#seed the random number generator
seed(1)
#generate univariate observations
data = 5 * randn(100) + 50
#normality test
stat, p = normaltest(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
#interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

 The **Anderson-Darling Test** is a statistical test that can be used to evaluate whether a data sample
 comes from one of among many known data samples. <br>
A feature of the Anderson-Darling test is that it **returns a list
 of critical values** rather than a single p-value. This can provide the basis for a more thorough
 interpretation of the result.<br>
 It takes as parameters the data sample and the name of the distribution to test it against.

What Does the Test Actually Do?
* It compares the actual data sample distribution to a theoretical distribution (default = normal).

* It calculates a test statistic that measures how far your data deviates from the normal distribution.

* It provides a set of critical values at different significance levels (e.g. 15%, 10%, 5%, 2.5%, 1%).

```python
# example of the anderson-darling test
from numpy.random import seed
from numpy.random import randn
from scipy.stats import anderson
# seed the random number generator
seed(1)
# generate univariate observations
data = 5 * randn(100) + 50
# normality test
result = anderson(data)
print('Statistic: %.3f' % result.statistic)
p = 0
for i in range(len(result.critical_values)):
  sl, cv = result.significance_level[i], result.critical_values[i]
  if result.statistic < result.critical_values[i]:
    print('%.3f: %.3f, data looks normal (fail to reject H0)' % (sl, cv))
  else:
    print('%.3f: %.3f, data does not look normal (reject H0)' % (sl, cv))
```

Running the example calculates the statistic on the test data set and prints the critical values.<br>
 Critical values in a statistical test are a range of pre-defined significance boundaries at which
 the H0 is failed to be rejected if the calculated statistic is less than the critical value.<br>
Rather than just a single p-value, the test returns a critical value for a range of di erent commonly
 used significance levels.<br> We can interpret the results by failing to reject the null hypothesis
 that the data is normal if the calculated test statistic is less than the critical value at a chosen
 signi cance level.<br> We can see that at each signi cance level, the test has found that the data
 follows a normal distribution.

# Make Data Normal:

 It is possible that your data does not look Gaussian or fails a normality test, but can be
 transformed to make it fit a Gaussian distribution. <br>This is more likely if you are familiar with
 the process that generated the observations and you believe it to be a Gaussian process, or
 the distribution looks almost Gaussian, except for some distortion.

Reasons why data sample is non-Gaussian: <br>
1. Size of the data sample is too small
2. Data resolution
3. Extreme values (outliers)

**The law of large numbers** states that

>  The more data that you collect, the more likely your
 data will be able to used to describe the underlying population distribution.



**1. Data Size:**

```python
#example on the small sample size
# histogram plot of a small sample
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot
# seed the random number generator
seed(1)
# generate a univariate data sample
data = 50 * randn(10) + 100
# histogram
pyplot.hist(data)
pyplot.show()
```

 Running the example creates a histogram plot of the data showing no clear Gaussian
 distribution, not even Gaussian-like.


```python
#increasing the sample size from 10 to 100
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot
# seed the random number generator
seed(1)
# generate a univariate data sample
data = 50 * randn(100) + 100
# histogram
pyplot.hist(data)
pyplot.show()
```

**2. Data Resolution:**
<bR> Data resolution refers to the level of detail or granularity at which data is captured, represented, or measured. <br>
For example:
High resolution: 3.74, 3.75, 3.76
<br>Low resolution: 3.7, 3.8, 3.8 (rounded or truncated)

 There may be many reasons why the
 resolution of the data is being modified prior to modeling, such as:
 * The configuration of the mechanism making the observation.
 * The data is passing through a quality-control process.
 * The resolution of the database used to store the data.

```python
# histogram plot of a low res sample
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot
# seed the random number generator
seed(1)
# generate a univariate data sample
data = randn(100)
# remove decimal component
data = data.round(0)
# histogram
pyplot.hist(data)
pyplot.show()
```

Even though the data comes from a normal distribution, the rounded result does not look Gaussian anymore.

This is the key message:


> Low-resolution data can hide the real shape of your distribution.



**3. Extreme Values:**
<br>  Extreme
 values could be present for a number of reasons, such as:

* Measurement error
* Missing data
* Data corruption
* Rare events
<br> In such cases, the extreme values could be identified and removed in order to make the
 distribution more Gaussian.

```python
# histogram plot of data with outliers
from numpy.random import seed
from numpy.random import randn
from numpy import zeros
from numpy import append
from matplotlib import pyplot
# seed the random number generator
seed(1)
# generate a univariate data sample
data = 5 * randn(100) + 10
# add extreme values
data = append(data, zeros(10))
# histogram
pyplot.hist(data)
pyplot.show()
```

 Extreme values can manifest in many ways. In addition to an abundance of rare events at the
 edge of the distribution, you may see a long tail on the distribution in one or both directions.

**Long Tails:**<br>
 In plots, this can make the distribution look like it is exponential, when in fact it might be
 Gaussian with an abundance of rare events in one direction. <br>
 You could use simple threshold
 values, perhaps based on the number of standard deviations from the mean, to identify and
 remove long tail values.

```python
# histogram plot of data with a long tail
from numpy.random import seed
from numpy.random import randn
from numpy.random import rand
from numpy import append
from matplotlib import pyplot
# seed the random number generator
seed(1)
# generate a univariate data sample
data = 5 * randn(100) + 10
tail = 10 + (rand(50) * 100)
# add long tail
data = append(data, tail)
# histogram
pyplot.hist(data)
pyplot.show()
```

 Running the example you can see how the long tail distorts the Gaussian distribution and
 makes it look almost exponential or perhaps even bimodal (two bumps).

We can use a simple threshold, such as a value of 25, on this dataset as a cuto and remove
 all observations higher than this threshold.<br>
  We did choose this threshold with prior knowledge
 of how the data sample was contrived, but you can imagine testing different thresholds on your
 own dataset and evaluating their effect.

```python
# histogram plot of data with a long tail
from numpy.random import seed
from numpy.random import randn
from numpy.random import rand
from numpy import append
from matplotlib import pyplot
# seed the random number generator
seed(1)
# generate a univariate data sample
data = 5 * randn(100) + 10
tail = 10 + (rand(10) * 100)
# add long tail
data = append(data, tail)
# trim values
data = [x for x in data if x < 25]
# histogram
pyplot.hist(data)
pyplot.show()
```

Running the code shows how this simple trimming of the long tail returns the data to a
 Gaussian distribution.

**Power Transforms:** <br>
The distribution of the data may be normal, but the data may require a transform in order to
 help expose it.<br>
 Examples: <br>
 1. Skewed data ->  this can be corrected by transforming the
 data via calculating the **square root** of the observations.
 2. Exponential data -> may look normal if the observations are transformed by taking the natural
 logarithm **bold text** of the values.


> NOTE: Data with this distribution is called **log-normal.**



```python
#log-normal distribution
from numpy.random import seed , randn
from numpy import exp
from matplotlib import pyplot
seed(1)
data = 5 * randn(100) + 50
#transform the data to be exponential
data = exp(data)
pyplot.hist(data)
pyplot.show()
```

**The Box-Cox method:**<br>
It is a
 data transform method that is able to perform a range of power transforms, including the log
 and the square root. <br>
 More than that,
 it can be configured to evaluate a suite of transforms automatically and select a best fit.


 The **boxcox()** SciPy function implements the Box-Cox method. It takes an argument, called
**lambda**, that controls the type of transform to perform. Below are some common values for
 lambda:

* lambda =-1.0 is a reciprocal transform.
* lambda =-0.5 is a reciprocal square root transform.
* lambda = 0.0 is a log transform.
* lambda = 0.5 is a square root transform.
* lambda = 1.0 is no transform

```python
from scipy.stats import boxcox
# power transform
data = boxcox(data, 0)
# histogram
pyplot.hist(data)
pyplot.show()
```

# Rank Correlation:

**Correlation vs Covariance:** <br>
Covariance and correlation both measure the relationship between two variables, but correlation standardizes the measure, making it easier to interpret and compare across different datasets. <br> **Covariance** indicates the **direction of the relationship** (positive or negative), while **correlation** indicates **both direction and strength** (ranging from -1 to +1).

When we do not
 know the distribution of the variables, we must use nonparametric rank correlation methods.

**Correlation for Gaussian distribution:** <br>
* Pearson's correlation

**Correlation for Non-Gaussian distribution**: <br>
* Rank correlation methods


 Rank correlation methods
 must be used. Rank correlation refers to methods that quantify the association between variables
 using the ordinal relationship between the values rather than the specific values.

 Because no distribution for the values is assumed,
 rank correlation methods are referred to as distribution-free correlation or nonparametric
 correlation.

Four examples of rank correlation methods are as follows:
* Spearmans Rank Correlation.
* Kendalls Rank Correlation.
* Goodman and Kruskals Rank Correlation.
* Somers Rank Correlation.

```python
# generate related variables
from numpy.random import rand
from numpy.random import seed
from matplotlib import pyplot
# seed random number generator
seed(1)
# prepare data
data1 = rand(1000) * 20
data2 = data1 + (rand(1000) * 10)
# plot
pyplot.scatter(data1, data2)
pyplot.show()
```

1.**Spearsman's Rank Correlation:** <br>This statistical method quantifies the degree to which ranked
 variables are associated by a monotonic function, meaning an increasing or decreasing relationship.

 The intuition for the Spearmans rank correlation is that it calculates a Pearsons correlation
 (e.g. a parametric measure of correlation) using the rank values instead of the real values.

 Spearmans rank correlation can be calculated in Python using the **spearmanr()**
 SciPy function.
 The function:
 *  Takes two real-valued samples as arguments.
 *  Returns both the correlation coefficient in the range between-1 and 1 and the p-value for interpreting the
 significance of the coefficient.


```python
# calculate the spearman's correlation between two variables
from numpy.random import rand
from numpy.random import seed
from scipy.stats import spearmanr
# seed random number generator
seed(1)
# prepare data
data1 = rand(1000) * 20
data2 = data1 + (rand(1000) * 10)
# calculate spearman's correlation
coef, p = spearmanr(data1, data2)
print('Spearmans correlation coefficient: %.3f' % coef)
# interpret the significance
alpha = 0.05
if p > alpha:
  print('Samples are uncorrelated (fail to reject H0) p=%.3f' % p)
else:
  print('Samples are correlated (reject H0) p=%.3f' % p)
```

2. **Kendall's Rank Correlation:** <br>
 The intuition for the test is that it calculates a
 normalized score for the number of matching or concordant rankings between the two samples.


The Kendall's rank correlation coefficient can be calculated in Python using the **kendalltau()** SciPy function.<br>
 The test takes
 the two data samples as arguments and returns the correlation coefficient and the p-value.

```python
# calculate the kendall's correlation between two variables
from numpy.random import rand
from numpy.random import seed
from scipy.stats import kendalltau
# seed random number generator
seed(1)
# prepare data
data1 = rand(1000) * 20
data2 = data1 + (rand(1000) * 10)
# calculate kendall's correlation
coef, p = kendalltau(data1, data2)
print('Kendall correlation coefficient: %.3f' % coef)
# interpret the significance
alpha = 0.05
if p > alpha:
  print('Samples are uncorrelated (fail to reject H0) p=%.3f' % p)
else:
  print('Samples are correlated (reject H0) p=%.3f' % p)
```
